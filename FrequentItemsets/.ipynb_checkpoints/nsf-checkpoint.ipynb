```python
import xml.etree.ElementTree as ET

def getIDCount(e):
    PIs = []
    try:
        elem = ET.fromstring(e[1])
        for n in elem.findall("./Award/Investigator/NSF_ID"):
            if (n.text != ''):
                PIs.append((n.text,1))
    except:
        PIs.append(('Exception',1))
    return PIs
getIDCount(s[0])
```

```python
tmp = raw_data.flatMap(getIDCount)
tmp.count()
```

```python
import xml.etree.ElementTree as ET

def getIDCount2(e):
    PIs = []
    try:
        elem = ET.fromstring(e[1])
        for n in elem.findall("./Award/Investigator/NSF_ID"):
            if (n.text != ''):
                PIs.append((n.text,1))
    except:
        PIs.append((e,0))
    return PIs
getIDCount(s[0])
```

```python
import xml.etree.ElementTree as ET

def getIDCount3(e):
    PIs = []
    try:
        elem = ET.fromstring(e[1])
    except:
        elem = ET.fromstring(e[1].split('<POR>',1)[0] + "</Award>\n</rootTag>\n")

    for n in elem.findall("./Award/Investigator/NSF_ID"):
            if (n.text != ''):
                PIs.append((n.text,1))
    return PIs
print(getIDCount3(s[0]))
print(getIDCount3(sb[0][0]))
```
